```python
name_depart = list(data_distrito["NOMBDEP"].unique())
```

```python
print(name_depart)
```

```python
name = 'AYACUCHO'
```

```python
depart_ayacucho = data_distrito[data_distrito["NOMBDEP"] == name]
```

```python
depart_ayacucho.plot(color = "green")
plt.show()
```

```python
depart_ayacucho.crs
```

```python
depart_ayacucho.head(3)
```

```python
depart_ayacucho.crs.name
```

```python
name_colm = list(depart_ayacucho.columns)
```

```python
print(name_colm)
```

```python
name_colm.remove("NOMBDIST")
name_colm.remove("NOMBPROV")
name_colm.remove("NOMBDEP")
name_colm.remove("geometry")
```

```python
print(name_colm)
```

```python
depart_ayacucho_pop = depart_ayacucho.drop(labels = name_colm, axis = 1)
depart_ayacucho_pop.head(3)
```

```python
depart_ayacucho_pop['centroid'] = depart_ayacucho_pop['geometry'].centroid
```

```python
depart_ayacucho_pop.columns
```

```python
%matplotlib inline
```

```python
depart_ayacucho_pop.head(3)
```

```python
fig, ax = plt.subplots(1, figsize=(22,8))


depart_ayacucho_pop.plot(column='NOMBDIST',
                         cmap='OrRd', 
                         k=20, 
                         legend= True,
                         legend_kwds={ 'title': 'Distritos de Ayacucho',
                             'bbox_to_anchor':(1.9,1),
                             'loc': 'upper center'
                        }, ax=ax)
                                     
        
                    

props = dict(boxstyle= 'round' , facecolor= 'linen' , alpha=1)

for point in  depart_ayacucho_pop.iterrows():
    ax.text(point[1]['centroid'].x,
            point[1]['centroid'].y,
            point[1]['NOMBDIST'],
            horizontalalignment='center',
            fontsize= 5,
            bbox=props)

ax.axis('on')
plt.title('Mapa distrital del departamento Ayacucho')
plt.xlabel('Este')
plt.ylabel('Norte')


outfp = "Mapa_ayacucho"
plt.savefig(outfp, dpi=500)
plt.show()
            
    

                         
```
